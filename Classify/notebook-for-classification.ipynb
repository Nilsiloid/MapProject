```python
# Specify the path to your images folder
images_folder = "/kaggle/input/trainingimages/images"

# Get a list of all image files in the folder
image_files = [f for f in os.listdir(images_folder) if f.endswith((".jpg", ".jpeg", ".png"))]

# Create or open a text file to save predictions
output_file_path = "predictions.txt"
with open(output_file_path, "w") as output_file:
    # Run the model on each image
    for image_file in image_files:
        image_path = os.path.join(images_folder, image_file)
        
        # Make predictions using the model
        predictions = model.predict(image_path=image_path)
        
        # Write the predictions to the output file
        output_file.write(f"Predictions for {image_file}: {predictions}\n")

print(f"Predictions saved to {output_file_path}")
```

# This saves map number and predicted class

```python
import re

# Specify the path to your input and output files
input_file_path = '/kaggle/input/prediction-for-accuracy/parsed_predictions.txt'
output_file_path = 'parsed_predictions.txt'

# Read the content of the input file
with open(input_file_path, 'r') as file:
    content = file.readlines()

# Modify each line and write to the output file
with open(output_file_path, 'w') as output_file:
    for line in content:
        # Extract the map number and the number 0 or 1
        match = re.match(r'map_(\d+)_png (\d)', line)
        if match:
            map_number, status = match.groups()

            # Write the modified content to the output file
            output_file.write(f"{map_number} {status}\n")

print(f"File modification completed. Output file: {output_file_path}")
```

# Calculating accuracy

```python
# Specify the paths to your input files
final_actual_path = '/kaggle/working/FinalActualPred.txt'
final_model_pred_path = '/kaggle/working/FinalModelPred.txt'

# Read the content of both files
with open(final_actual_path, 'r') as file_actual, open(final_model_pred_path, 'r') as file_model_pred:
    lines_actual = file_actual.readlines()
    lines_model_pred = file_model_pred.readlines()

# Check the number of lines in both files
num_lines_actual = len(lines_actual)
num_lines_model_pred = len(lines_model_pred)

# Ensure both files have the same number of lines
if num_lines_actual != num_lines_model_pred:
    print("Error: The number of lines in the two files is not the same.")
else:
    # Count the entries where the second number is the same in both files
    same_count = sum(1 for actual, model_pred in zip(lines_actual, lines_model_pred) if actual.split()[1] == model_pred.split()[1])

    print(f"Total number of entries where the second number is the same: {same_count}")
```
